```python
import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'Order Date' is datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)

# Group by year and sum sales
yearly_sales = df.groupby(df['Order Date'].dt.year)['Sales'].sum().reset_index()

# Rename columns
yearly_sales.columns = ['Year', 'Total Sales']

# Plot
plt.plot(yearly_sales['Year'], yearly_sales['Total Sales'], marker='o', linestyle='-')
plt.title('Yearly Sales')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()

```

```python
import pandas as pd

# Ensure 'Order Date' is datetime
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)

# Filter for 2016
year_sales = df[df['Order Date'].dt.year == 2016]
print("2016 rows:", len(year_sales))

# Use new frequency alias 'QE-DEC'
if not year_sales.empty:
    quarterly_sales = year_sales.resample('QE-DEC', on='Order Date')['Sales'].sum().reset_index()
    quarterly_sales = quarterly_sales.rename(columns={'Order Date': 'Quarter', 'Sales': 'Total Sales'})
    print(quarterly_sales)
else:
    print("No sales data found for 2016.")



```

```python
plt.plot(quarterly_sales['Quarter'], quarterly_sales['Total Sales'], marker = 'o', linestyle = '--')

plt.tight_layout()
plt.xticks(rotation=75)
plt.show()
```

```python
import pandas as pd

df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce', dayfirst=True)
yearly_sales = df[df['Order Date'].dt.year == 2016]

if not yearly_sales.empty:
    monthly_sales = yearly_sales.resample('ME', on='Order Date')['Sales'].sum().reset_index()
    monthly_sales = monthly_sales.rename(columns={'Order Date': 'Month', 'Sales': 'Total Monthly Sales'})
    print(monthly_sales)
else:
    print("No sales data found for 2016.")

```

```python
import matplotlib.pyplot as plt

plt.plot(monthly_sales['Month'], monthly_sales['Total Monthly Sales'], marker='o', linestyle='--')
plt.title('Monthly Sales for 2016')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```
